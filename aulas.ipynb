INTRODUÇÃO A LÓGICA DE PROGRAMAÇÃO

O que é lógica de programação?

Buscando o significado das palavras no contexto geral:

°Lógica - Forma de organizar o pensamento pra resolver problemas passo a passo, de maneira clara e coerente.

°Programação - Processo de escrever instruções (código) para que o computador execute tarefas.

Seguindo o contexto de lógica e programação
Há uma maneira que nós (programadores) podemos ensinar a máquina a seguir um passo a passo,
chamamos isso de "Lógica de programação".
```
Vamos usar um exemplo comum para a sequência de execução de tarefas:
Ação "Fazer Miojo"
1º Passo: Ferver água
2º Passo: Colocar o macarrão
3º Passo: Esperar 3 minutos
4º Passo: Temperar

Objetivo "Comer miojo pronto".
```

O que é um algoritmo?

Algoritmo são as etapas que predefinimos para a execução da atividade.
Aproveitando o exemplo anterior da preparação do miojo.
Iremos realizar mais ações dentro de cada etapa:
Atividade: Fazer miojo
Pegar miojo no armário;
Ler modo de preparo;
```
1º Passo: Ferver água
Pegar panela para por água;
Colocar a quantidade de água dita na embalagem;
Ligar boca do fogão;
Colocar panela com água na boca acesa;
Aguardar a água ferver.
```
```
2º Passo: Colocar o macarrão
Abrir pacote do Miojo;
Separar sache de tempero;
TIrar miojo da embalagem;
Colocar miojo na panela após água começar a ferver.
```
```
3º Passo: Esperar 3 minutos
Aguardar 3 minutos com o Miojo fora do pacote dentro da água fervendo
```
```
4º Passo: Temperar
Desligar boca do fogão;
Pegar pacote de tempero que veio com o Miojo;
Abrir e despejar tempero na água com miojo;
Pegar colher na gaveta;
Misturar água com miojo e tempero com a colher;
Retirar miojo pronto da panela;
(Opicional)
Coar miojo para retirar o exesso de água e tempero;
Adicionar miojo em prato e ou pote;
Pegar garfo;
Utilizar garfo para comer miojo no prato.
```

Percebem o quão extensa uma atividade tão simples se torna quando destrinchamos
o algoritmo dela?

Podem ser atividades simples, porém levam uma sequência de ações.


Quando nos referimos a algoritmo na programação, fica mais simples, por conta
que ordenamos a máquina a seguir os passos de maneira guiada, facilitando na execução e lógica.

# Programação em Python

As variáveis são como caixas que guardam informações para serem usadas dentro da lógica do programa.

```
Variáveis (guardar informação)

    As variáveis nada mais são que formas as quais levamos as informações para dentro da nossa lógica.

    Dentre as variáveis existem:
    int - Usada para definir números que pertencem aos reais (Números inteiros)
    float - Usado para definir números que pertencem aos irreais (Números com valor decimal)
    string - Usado para definir valor de tipo texto
    bool - valores lógicos (True ou False)
```

Vamos a programação **!!!**

````python
idade = 20 # entrada INT
altura = 1.80 # entrada FLOAT
nome = "Lucas" # entrada STRING
programa = True # entrada BOOLEANA

print("Nome:", nome)
print("Idade:", idade)
print("Altura:", altura)
print("Gosta de programação?", programa)
````
